```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Dataset**

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:



*   distplot/histplot
*   jointplot
*   pairplot
*   rugplot
*   kdeplot
*   displot



# **1. distplot / `histplot`**



*   Shows frequency distribution
*   kde=True adds smooth density curve



```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20, kde=True)

plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=20, kde=True)

plt.show()
```

# **2. `jointplot`**



*   Shows relationship between two variables
*   Includes scatter + distribution



```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')    # scatter
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')      # regression
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')        # hex – useful for large data
```

# **3. `pairplot`**



*   Shows pairwise relationships across all numerical columns




```python
sns.pairplot(df, hue='size', palette='rainbow')
```

```python
sns.pairplot(df[['total_bill','tip','size']], hue='size')      # Variation (only selected columns)
```

# **4. `rugplot`**



*   Shows distribution as small ticks on axis



```python
sns.rugplot(df['tip'])
```

```python
sns.kdeplot(df['tip'])
sns.rugplot(df['tip'])
```

# **5. `kdeplot`**



*   Smooth probability density curve



```python
sns.kdeplot(df['total_bill'])
```

**Filled KDE**

```python
sns.kdeplot(df['total_bill'], fill=True)
```

# **6. displot**



*   Figure-level version of histplot
*   More flexible



```python
sns.displot(df['total_bill'], kde=True, bins=20)
```

**Multiple distributions**

```python
sns.displot(data=df, x='total_bill', col='time', kde=True)
```

# **7. Comparing distributions**

```python
sns.histplot(data=df, x='total_bill', hue='sex', kde=True)     # multiple categories
```

```python
sns.histplot(data=df, x='total_bill', hue='smoker', multiple='stack')       # stacked
```
