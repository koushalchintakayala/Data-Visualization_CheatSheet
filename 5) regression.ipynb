````python
import seaborn as sns
import matplotlib.pyplot as plt
````

# **lmplot()**
regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

````python
tips = sns.load_dataset("tips")
tips
````

**Basic lmplot()**

````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips)
````

**Using hue**

````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           hue = 'sex')
````

**Change Color Palette**

`palette` changes graph colors

Some popular palettes:

```
'coolwarm'
'viridis'
'Set1'
'deep'
```



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           hue = 'sex',
           palette = 'coolwarm')
````

**Change Marker Style**

*   `markers` changes marker shapes
*   `scatter_kws` customizes scatter points
*   `s` means marker size

````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           hue = 'sex',
           palette = 'coolwarm',
           markers = ['o','v'],
           scatter_kws = {'s':50})
````

# **Additional Important Things to Learn in `lmplot()`**

# 1) Change Height and Aspect Ratio



*   `height` → controls graph height
*   `aspect` → controls width ratio



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           height = 6,
           aspect = 2)
````

# 2) Remove Regression Line




*   `fit_reg = False` removes regression line
*   only scatter points will be shown



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           fit_reg = False)
````

# 3) Different Columns using col

*   creates separate graphs column-wise


````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           col = 'sex')
````

# 4) Multiple Rows using row



*   creates graphs row-wise



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           row = 'smoker')
````

# 5) Using row and col Together



*   useful for comparing multiple categories together



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           row = 'smoker',
           col = 'sex')
````

# 6) Change Marker Size and Transparency




*   `s` → marker size
*   `alpha` → transparency



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           scatter_kws = {'s':100,'alpha':0.5})
````

# 7) Change Regression Line Color and Width

*   `line_kws` customizes regression line



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           line_kws = {'color':'red',
                        'linewidth':3})
````

# 8) Polynomial Regression



*   `order = 2` creates curved regression line



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           order = 2)
````

# 9) Remove Confidence Interval



*   removes shaded confidence interval region



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           ci = None)
````

# 10) Logistic Regression



*   used for classification type visualization



````python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

sns.lmplot(x = 'total_bill',
           y = 'size',
           data = tips,
           logistic = True)
````

# 11) Robust Regression

*   handles outliers better



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           robust = True)
````

# 12) Lowess Smoothing
*   creates smooth curve instead of straight line



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           lowess = True)
````

# 13) Change Figure Size Completely



*   bigger graph gives better visualization



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           height = 8,
           aspect = 1.5)
````

# 14) Customize Scatter Point Color

````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           scatter_kws = {'color':'green'})
````

# 15) Customize Regression Line Separately

````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           line_kws = {'color':'black'})
````

# 16) Disable Scatter Points

*   only regression line will appear


````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           scatter = False)
````

# 17) Truncate Regression Line

*   regression line stays within data limits



````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           truncate = True)
````

# 18) Jittering Data Points



*   helps avoid overlapping points



````python
sns.lmplot(x = 'size',
           y = 'tip',
           data = tips,
           x_jitter = 0.2)
````

# **Important Parameters of lmplot()**

| Parameter   | Purpose                   |
| ----------- | ------------------------- |
| x           | x-axis column             |
| y           | y-axis column             |
| data        | dataset                   |
| hue         | category grouping         |
| palette     | color theme               |
| markers     | marker styles             |
| scatter_kws | scatter customization     |
| line_kws    | line customization        |
| fit_reg     | show/hide regression line |
| height      | graph height              |
| aspect      | graph width ratio         |
| row         | row-wise graphs           |
| col         | column-wise graphs        |
| ci          | confidence interval       |
| order       | polynomial regression     |
| logistic    | logistic regression       |
| robust      | robust regression         |
| lowess      | smooth regression         |
| scatter     | show/hide scatter points  |
| truncate    | limit regression line     |
| x_jitter    | spread overlapping points |




# **Difference Between regplot() and lmplot()**

| regplot()                 | lmplot()                |
| ------------------------- | ----------------------- |
| axes-level function       | figure-level function   |
| works on matplotlib axes  | works on FacetGrid      |
| simpler                   | advanced                |
| no row/col support easily | supports row and col    |
| used for single plot      | used for multiple plots |


````python
sns.lmplot(x = 'total_bill',
           y = 'tip',
           data = tips,
           hue = 'sex',
           col = 'smoker',
           palette = 'coolwarm',
           markers = ['o','v'],
           height = 5,
           aspect = 1,
           scatter_kws = {'s':80,'alpha':0.7},
           line_kws = {'linewidth':2},
           ci = None)

plt.show()
````
