```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Matplotlib Basics**

```python
x = np.linspace(0,5,11)
x
y = x **3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("Chacha")
plt.ylabel("Chachi")
```

**Multiple Lines in Same Plot**

```python
plt.plot(x, y)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.show()
```

**Adding Legend**

```python
plt.plot(x, y, label="x^3")
plt.plot(x, x**2, label="x^2")

plt.legend()
plt.show()
```

**Grid**

```python
plt.plot(x, y)
plt.grid(True)
plt.show()
```

**Axis Limits**

```python
plt.plot(x, y)

plt.xlim(0,3)
plt.ylim(0,30)

plt.show()
```

# **Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(x,y)

plt.subplot(2,2,3)
plt.plot(y,x)

plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

# **Subplots using Object-Oriented**

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x, y)
ax[0][1].plot(y, x)
ax[1][0].plot(x, x**2)
ax[1][1].plot(x, x**3)

plt.tight_layout()
```

**Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,8), dpi=100)

axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)

axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

**Tight Layout**

```python
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("Graph")

plt.tight_layout()
```

**Line Width + Style**

```python
xaxis = np.linspace(0,5,11)

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(xaxis, xaxis+1, color="red", linewidth=0.25)
ax.plot(xaxis, xaxis+2, color="red", linewidth=0.50)
ax.plot(xaxis, xaxis+3, color="red", linewidth=1.00)
ax.plot(xaxis, xaxis+4, color="red", linewidth=2.00)

ax.plot(xaxis, xaxis+5, color="green", lw=3, linestyle='-')
ax.plot(xaxis, xaxis+6, color="green", lw=3, ls='-.')
ax.plot(xaxis, xaxis+7, color="green", lw=3, ls=':')
```

**Custom Dash**

```python
line, = ax.plot(xaxis, xaxis+8, color="black", lw=1.5)

line.set_dashes([5,50,15,50])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.arange(0, 6)

fig, ax = plt.subplots()

line, = ax.plot(xaxis, xaxis + 8, color="black", lw=1.5)

line.set_dashes([5, 50, 15, 50])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.arange(0, 10)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(xaxis, xaxis+13, color="purple", marker='o', markersize=2)

ax.plot(xaxis, xaxis+14, color="purple", marker='o', markersize=4)

ax.plot(xaxis, xaxis+15, color="purple", marker='o',
        markersize=8, markerfacecolor="red")

ax.plot(xaxis, xaxis+16, color="purple", marker='s',
        markersize=8, markerfacecolor="yellow",
        markeredgewidth=3, markeredgecolor="green")

plt.show()
```

**Fill Area**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)
y = x * 2

plt.plot(x, y)

plt.fill_between(x, y)

plt.show()
```

```python
plt.fill_between(x, y, color='skyblue')

plt.show()
```

```python
plt.fill_between(x, y, color='green', alpha=0.4)

plt.show()
```

**Error Bars**

```python
error = np.random.rand(len(x))

plt.errorbar(x, y, yerr=error)
plt.show()
```

**Log Scale**

```python
plt.plot(x, y)
plt.yscale('log')
plt.show()
```

**Twin Axis**

```python
fig, ax1 = plt.subplots()

ax1.plot(x, y, color="blue")

ax2 = ax1.twinx()
ax2.plot(x, x**2, color="red")

plt.show()
```

**Annotations**

```python
plt.plot(x, y)

plt.annotate("Peak",
             xy=(5,125),
             xytext=(3,100),
             arrowprops=dict(facecolor='black'))

plt.show()
```

**Style Themes**

```python
plt.style.use('ggplot')

plt.plot(x, y)
plt.show()
```

**Quick Notes**

*   `plt` → quick plotting
*   `fig, ax` → preferred (object-oriented)
*   `figsize` → controls size
*   `dpi` → resolution
*   `tight_layout()` → avoid overlapping
*   `legend()` → label lines
*   `twinx()` → dual axis
*   `annotate()` → highlight points
