````python
import seaborn as sns
import matplotlib.pyplot as plt
````

# **Heat map**

````python
flights = sns.load_dataset('flights')
flights
````

````python
tips = sns.load_dataset("tips")
tips
````

# **Correlation Heat Map**

**Selecting Numerical Columns**

````python
tipscorr  = tips[['total_bill','tip','size']]
tipscorr
````

**Finding Correlation**

````python
tipscorr.corr()
````

**Basic Heatmap**

````python
sns.heatmap(tipscorr.corr(),annot=True)
plt.show()
````

# **Important Heatmap Parameters**
**`cmap`**

Changes color style.

````python
sns.heatmap(tipscorr.corr(),annot=True,cmap='coolwarm')
plt.show()
````

**Other useful colormaps:**


```
cmap='viridis'
cmap='magma'
cmap='YlGnBu'
cmap='Blues'
```



**`linewidths` and `linecolor`**

Adds borders between cells.

````python
sns.heatmap(
    tipscorr.corr(),
    annot=True,
    linewidths=2,
    linecolor='black'
)
plt.show()
````

**`vmin` and `vmax`**

Controls heat intensity range.

````python
sns.heatmap(
    tipscorr.corr(),
    annot=True,
    vmin=-1,
    vmax=1
)
plt.show()
````

**`fmt`**

Changes decimal format.

````python
sns.heatmap(
    tipscorr.corr(),
    annot=True,
    fmt='.1f'
)
plt.show()
````

**`cbar`**

Removes side color bar.

````python
sns.heatmap(
    tipscorr.corr(),
    annot=True,
    cbar=False
)
plt.show()
````

**`square`**

Makes cells square shaped.

````python
sns.heatmap(
    tipscorr.corr(),
    annot=True,
    square=True
)
plt.show()
````

**`annot_kws`**

Changes annotation text style.

````python
sns.heatmap(
    tipscorr.corr(),
    annot=True,
    annot_kws={'size':12}
)
plt.show()
````

# **Cluster Map**

**Basic Cluster Map**

````python
sns.clustermap(tipscorr.corr())
plt.show()
````

**Cluster Map with cmap**

````python
sns.clustermap(
    tipscorr.corr(),
    cmap='coolwarm',
    annot=True
)
plt.show()
````

**Standard Scaling**

````python
sns.clustermap(
    tipscorr.corr(),
    standard_scale=1
)
plt.show()
````

# **Pivot Table Heat Map**
**Creating Pivot Table**

````python
pvtflight = flights.pivot_table(
    values='passengers',
    index='month',
    columns='year'
)

pvtflight
````

**Basic Heatmap**

````python
sns.heatmap(pvtflight)
plt.show()
````

**Heatmap with cmap**

````python
sns.heatmap(
    pvtflight,
    cmap='coolwarm'
)
plt.show()
````

**Heatmap with line borders**

````python
sns.heatmap(
    pvtflight,
    cmap='viridis',
    linewidths=1
)
plt.show()
````

**Display Passenger Values**

````python
sns.heatmap(
    pvtflight,
    annot=True,
    fmt='.0f'
)
plt.show()
````

**Center Parameter**

Useful for balancing colors.

````python
sns.heatmap(
    pvtflight,
    center=300
)
plt.show()
````

# **Cluster Map with Pivot Table**

````python
sns.clustermap(pvtflight)
plt.show()
````

**Cluster Map with Standard Scaling**

````python
sns.clustermap(
    pvtflight,
    standard_scale=1
)
plt.show()
````

**Cluster Map with z_score**

````python
sns.clustermap(
    pvtflight,
    z_score=1
)
plt.show()
````

# **Figure Size Customization**

````python
plt.figure(figsize=(10,6))

sns.heatmap(
    tipscorr.corr(),
    annot=True
)

plt.show()
````

# **Removing Annotation**

````python
sns.heatmap(tipscorr.corr())
plt.show()
````

# **Heatmap with Custom Tick Rotation**

````python
sns.heatmap(pvtflight)

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
````

# **Missing Values Heatmap**

Very important for interviews and EDA.

````python
sns.heatmap(tips.isnull())
plt.show()
````

# **Sorting Before Heatmap**

````python
sortedcorr = tipscorr.corr().sort_values(by='tip')

sns.heatmap(sortedcorr,annot=True)

plt.show()
````

# **Complete Advanced Example**

````python
plt.figure(figsize=(12,7))

sns.heatmap(
    pvtflight,
    cmap='coolwarm',
    annot=True,
    fmt='.0f',
    linewidths=1,
    linecolor='white'
)

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
````

# **Real Interview Style Example**
**Finding Strongly Related Features**

````python
corrdata = tipscorr.corr()

sns.heatmap(
    corrdata,
    annot=True,
    cmap='coolwarm',
    vmin=-1,
    vmax=1
)

plt.show()
````

**Groupby + Heatmap**

````python
groupdata = tips.groupby('day')[['total_bill','tip']].mean()

sns.heatmap(
    groupdata,
    annot=True,
    cmap='YlGnBu'
)

plt.show()
````
