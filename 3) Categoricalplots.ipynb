```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Categorical Data Plots**

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:



*   boxplot
*   violinplot
*   stripplot
*   swarmplot
*   barplot
*   countplot
*   catplot
*   Figure Size
*    





Let's go through examples of each!

```python
df = sns.load_dataset("tips")
df
```

# **`countplot`**

Used to count occurrences of categorical values.

```python
sns.countplot(x = df['sex'], hue = df['smoker'])
```

**Simple countplot**

```python
sns.countplot(x = df['day'])
```

**Using palette**

```python
sns.countplot(x = df['day'], palette = 'Set2')
```

**Changing order**

```python
sns.countplot(
    x = df['day'],
    order = ['Thur','Fri','Sat','Sun']
)
```

**Horizontal countplot**

```python
sns.countplot(y = df['day'])
```

# **`barplot`**

Used to show aggregate values like mean, sum, median etc.

Default estimator = mean

```python
sns.barplot(x = df['sex'], y = df['total_bill'])
```

You can change the estimator object to your own function.



*   Mean (default): `np.mean`
*   Median: `np.median`
*   Sum: `np.sum`
*   Count: `len`
*   Standard Deviation: `np.std`
*   Minimum: `np.min`
*   Maximum: `np.max`









**Sum estimator**

```python
import numpy as np
import seaborn as sns

sns.barplot(
    x = df['sex'],
    y = df['tip'],
    estimator = np.sum
)
```

**Median estimator**

```python
sns.barplot(
    x = df['day'],
    y = df['total_bill'],
    estimator = np.median
)
```

**With hue**

```python
sns.barplot(
    x = df['day'],
    y = df['total_bill'],
    hue = df['sex']
)
```

**Horizontal barplot**

```python
sns.barplot(
    x = df['total_bill'],
    y = df['day']
)
```

# **`boxplot`**

Boxplots are used to show the distribution of categorical data.

A box plot (or box-and-whisker plot) shows:


*   median
*   quartiles
*   spread
*   outliers

```python
sns.boxplot(
    x = 'tip',
    y = 'day',
    data = df,
    hue = 'day',
    palette = 'rainbow',
    legend = False
)
```

**Standard boxplot**

```python
sns.boxplot(
    x = 'day',
    y = 'total_bill',
    data = df
)
```

**With hue**

```python
sns.boxplot(
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    data = df
)
```

**Horizontal boxplot**

```python
sns.boxplot(
    x = 'total_bill',
    y = 'day',
    data = df
)
```

**Show mean**

```python
sns.boxplot(
    x = 'day',
    y = 'total_bill',
    showmeans = True,
    data = df
)
```

# **`violinplot`**

Violinplots are used to show the distribution and density of data.

It combines:


*   boxplot
*   density plot



```python
sns.violinplot(
    x = 'day',
    y = 'total_bill',
    data = df
)
```

**With hue**

```python
sns.violinplot(
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    data = df
)
```

**Split violinplot**

```python
sns.violinplot(
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    split = True,
    data = df
)
```

**Inner quartiles**

```python
sns.violinplot(
    x = 'day',
    y = 'total_bill',
    inner = 'quartile',
    data = df
)
```

**Horizontal violinplot**

```python
sns.violinplot(
    x = 'total_bill',
    y = 'day',
    data = df
)
```

# **`stripplot`**

The stripplot will draw a scatterplot where one variable is categorical.

It is useful for showing all observations.

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    data = df
)
```

**Using jitter**

Jitter helps avoid overlapping.

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    jitter = True,
    data = df
)
```

**With hue**

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    data = df
)
```

**Horizontal stripplot**

```python
sns.stripplot(
    x = 'total_bill',
    y = 'day',
    data = df
)
```

# **`swarmplot`**

The swarmplot is similar to stripplot(), but the points are adjusted so they don’t overlap.

```python
sns.swarmplot(
    x = 'day',
    y = 'total_bill',
    data = df
)
```

**With hue**

```python
sns.swarmplot(
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    data = df
)
```

**Reduce marker size**

Useful when warnings appear.

```python
sns.swarmplot(
    x = 'day',
    y = 'total_bill',
    size = 3,
    data = df
)
```

**Horizontal swarmplot**

```python
sns.swarmplot(
    x = 'total_bill',
    y = 'day',
    data = df
)
```

# **Combining violinplot and swarmplot**

Very useful in projects and interviews.

```python
sns.violinplot(
    x = 'tip',
    y = 'day',
    data = df
)

sns.swarmplot(
    x = 'tip',
    y = 'day',
    color = 'black',
    size = 3,
    data = df
)
```

# **`catplot`**

High-level seaborn categorical plot.

Can create:


*   boxplot
*   violinplot
*   stripplot
*   swarmplot
*   barplot
*   countplot

  using one function.



**Boxplot using catplot**

```python
sns.catplot(
    x = 'day',
    y = 'total_bill',
    kind = 'box',
    data = df
)
```

**Violinplot using catplot**

```python
sns.catplot(
    x = 'day',
    y = 'total_bill',
    kind = 'violin',
    data = df
)
```

**Countplot using catplot**

```python
sns.catplot(
    x = 'day',
    kind = 'count',
    data = df
)
```

**Barplot using catplot**

```python
sns.catplot(
    x = 'day',
    y = 'total_bill',
    kind = 'bar',
    data = df
)
```

# **Important Interview Concepts**

# Difference Between countplot and barplot
**`countplot`**

*   counts occurrences
*   no y value needed



```python
sns.countplot(x = df['day'])
```

**`barplot`**


*   calculates aggregate values
*   needs x and y





```python
sns.barplot(
    x = df['day'],
    y = df['tip']
)
```

# Difference Between boxplot and violinplot
**`boxplot`**
*   shows quartiles
*   easy to detect outliers

**`violinplot`**
*   shows complete distribution shape
*   shows density



# Difference Between stripplot and swarmplot
**`stripplot`**


*   points may overlap

**`swarmplot`**


*   avoids overlap automatically
